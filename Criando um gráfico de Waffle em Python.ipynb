**Criando um Gráfico de Waffle no Python.**

Essa é uma maneira simples e muito mais clara de apresentarmos os dados pesquisados.


```
#instalando o pywaffle
!pip install pywaffle -q
```

Vamos usar como exemplo de representação, uma pesquisa do jornal **El País** sobre o número de contaminados no **Brasil** por covid-19. Onde os dados atualizados no dia 17 de Agosto de 2020, demonstram que o País registra 3.359.570 casos confirmados


```
#importar as bibliotecas necessárias
import matplotlib.pyplot as plt
from pywaffle import Waffle

#exemplo de Waffle
fig = plt.figure(
                FigureClass=Waffle,
                rows=10,
                columns=27,
                values={'Distrito Federal':138735,
                        'Acre':22605,
                        'Alagoas':72938,
                        'Amapá':39703,
                        'Amazonas':112263,
                        'Bahia':217115,
                        'Ceará':198202,
                        'Espírito Santo':100859,
                        'Goiás':102665,
                        'Maranhão':137408,
                        'Mato Grosso':74509,
                        'Mato Grosso do Sul':37425,
                        'Minas Gerais':175715,
                        'Pará':178375,
                        'Paraiba':96223,
                        'Paraná':178375,
                        'Pernambuco':113183,
                        'Piauí':66315,
                        'Rio de Janeiro':194651,
                        'Rio Grande do Norte':57898,
                        'Rio Grande do Sul':98007,
                        'Rondônia':48232,
                        'Roraima':39665,
                        'Santa Catarina':121666,
                        'São Paulo':702665,
                        'Sergipe':68313,
                        'Tocantins':37856},           
               
                                
                legend={'loc': 'upper left',
                        'bbox_to_anchor': (1.1, 1)}
                )
fig.set_tight_layout(False)
plt.show()
```
