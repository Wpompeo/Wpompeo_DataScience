**Criando um Gráfico de Waffle no Python.**

Essa é uma maneira simples e muito mais clara de apresentarmos os dados pesquisados.

Vamos usar como exemplo de representação, dados extraidos do site: **https://covid.saude.gov.br/** no dia 19 de Agosto de 2020. Onde constam os seguintes números: 3.407.354 casos confirmados e 109.888 óbitos confirmados.


```
#importar as bibliotecas necessárias
import matplotlib.pyplot as plt
from pywaffle import Waffle

#exemplo de Waffle
fig = plt.figure(
                FigureClass=Waffle,
                rows=5,
                columns=10,
                values={'Brasil':3407354,
                        'Mortes':109888},           
               
                                
                legend={'loc': 'upper left',
                        'bbox_to_anchor': (1.1, 1)}
                )
fig.set_tight_layout(False)
plt.show()
```

Abaixo, temos a representação gráfica em barras, com os números de casos confirmados por Estados.
