**Criando um Gráfico de Waffle no Python.**

Essa é uma maneira simples e muito mais clara de apresentarmos os dados pesquisados.

Vamos usar como exemplo de representação, uma pesquisa do jornal **El País** sobre o número de contaminados no **Brasil** por covid-19. Onde os dados atualizados no dia 17 de Agosto de 2020, demonstram que o País registra 3.359.570 casos confirmados


```
#importar as bibliotecas necessárias
import matplotlib.pyplot as plt
from pywaffle import Waffle

#exemplo de Waffle
fig = plt.figure(
                FigureClass=Waffle,
                rows=5,
                columns=10,
                values={'Brasil':3411872,
                        'Mortes':110019},           
               
                                
                legend={'loc': 'upper left',
                        'bbox_to_anchor': (1.1, 1)}
                )
fig.set_tight_layout(False)
plt.show()
```

Este é apenas um inicio de estudo gráfico, aos poucos estarei melhorando seu layout.


```
#Visualização por estados
import matplotlib.pyplot as plt

x1 = ['DF','AC','AL','AP','AM','BA','CE','ES','GO','MA','MT',
      'MS','MG','PA','PB', 'PR','PE','PI','RJ','RN',
      'RS','RO','RR','SC','SP','SE','TO']                       
                      
y1 = [138735,22605,72938,39703,112263,217115,198202,100859,
      102665,137408,74509,37425,175715,178375,96223,178375,
      113183,66315,194651,57898,98007,48232,39665,121666,702665,68313,37856]
titulo = "Mortalidade COVID-19 Brasil"
eixox = "Estados"
eixoy = "N° de mortos"

plt.title(titulo, color= 'Black')
plt.xlabel(eixox, color= 'Black')
plt.ylabel(eixoy, color= 'Black')

plt.bar(x1, y1, color= 'black')
plt.plot(x1, y1, color='red', lineStyle='--',marker="*")
plt.rcParams['figure.figsize'] = (10,4) #define tamanho gráfico
plt.show()
```
